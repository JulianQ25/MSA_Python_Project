                   team, wins, losses, and win % for each team 
                   under a division header.You should round 
                   the percentage of games won to two 
                   places after the decimal.
#### INPUT:      
                   Reads mlb-games-2022.csv file that converts 
                   the file into a useable dataframe with Pandas.
#### PROCESS:    
                   Add headers to newly created dataframe.
                   Create and add win percentages to dataframe.
                   Group divisions then order divsion by ascending and
                   win percentages by descending order.
                   Rank each division by team win percentages.
                   Then save dataframe to new .csv file.
#### OUTPUT:

```python
import pandas as pd

#Input
def read_file():
  """
  This function reads the MLB
  .csv file using pandas and
  adds columns to the .csv file
  """
  mlb_file=pd.read_csv("mlb-games-2022.csv") #Reads CSV file using pandas and converts it to a dataframe
  headerList=['Teams','Wins','Losses','Divisions'] #creates a list of headers for the dataframe
  mlb_file.to_csv("mlb-games-2022.csv",header=headerList,index=False) #Adds headers to dataframe
  return mlb_file

#Process
def add_win_percentage(mlb_file):
  """
  This function is creating a new column
  to store the data for win
  percentages for each team
  """
  total_games=(mlb_file['Wins']+mlb_file['Losses']) #Calculates total games played for each team
  win_percentage=round((mlb_file['Wins']/total_games)*100,2) #Calculates total win percentage for each team
  mlb_file.insert(3,'Win Percentage',win_percentage) # Adds new column that contains win percentage to the dataframe
  return mlb_file

def group_divisions(mlb_file):
  """
  Moves Division column to first index
  then groups the data by
  divsions in ascending order
  """
  column_move=mlb_file.pop("Divisions") #Removes divisions from the dataframe
  mlb_file.insert(0,"Divisions",column_move) #inserts divisions back into index 0 so it is at the front of the dataframe
  mlb_file=mlb_file.sort_values(by=['Divisions','Win Percentage'],ascending=[True, False]) #Sorts dataframe by divisions first then win percentage
  return mlb_file

def rank_division(mlb_file):
  """
  Creates a new column called rank,
  this will create a ranking for
  each divsion in the mlb dataframe
  """
  mlb_file=mlb_file.reset_index(drop=True) #Reset file index
  mlb_file['Rank'] = mlb_file.groupby('Divisions').cumcount() + 1 #groups by divisions then adds rank starting at 1 then adding +1. Once a new division populates the rank starts back up from 1
  column_move1=mlb_file.pop("Rank") #removes rank column from dataframe
  mlb_file.insert(0,"Rank",column_move1) #inserts rank back into index 0 so it is at the front of the dataframe
  mlb_file.reset_index(drop=True) #Reset file index
  return mlb_file

# Output
def main():
    """
    Calls the functions
    and prints results.
    """
    mlb_data=read_file()
    print(mlb_data) #run the code with the print statement and it will work then you can comment it out
    mlb_data=add_win_percentage(mlb_data)
    mlb_data=group_divisions(mlb_data)
    mlb_data=rank_division(mlb_data)
    print('\t\t'"These are the standings for the MLB 2022 Season:",'\n')
    print(mlb_data)
    mlb_data.to_csv('mlb-division-standings-2022.csv',index=False)#saves the dataframe to a .csv file

if __name__ == "__main__":
    main()
```
